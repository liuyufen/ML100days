```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
#畫出完整的sin圖形
x = np.arange(0,3*np.pi,0.1)
y_sin = np.sin(x)
plt.plot(x,y_sin)
plt.show()
```

```python
#1畫出完整的cos圖檔並儲存成圖形檔
x = np.arange(0,3*np.pi,0.1)
y_cos = np.cos(x)
plt.plot(x,y_cos)
plt.savefig("cos.png",dpi=300,format="png")
```

```python
#2給定散點圖為紅色
n = 1024
x = np.random.normal(0,1,n)
y = np.random.normal(0,1,n)
plt.scatter(x,y,c = 'r')
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.show()
```

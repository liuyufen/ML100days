```python
sns.set_style('whitegrid')
```

```python
rs = np.random.RandomState(0)
x = rs.normal(2,1,75)
y = 2 + 1.5 * x + rs.normal(0,2,75)
```

```python
#繪製x變數的直方圖
sns.distplot(x)
```

```python
#繪製y變數的直方圖
sns.distplot(y)
```

```python
#繪製模型殘值圖
sns.residplot(x=x,y=y,lowess=True,color='r')
```

```python
#當圖形有KDE值，則Y軸數值為內核密度函數下的估計值，可視為不同類別分組bin間的相對比較值。
#當圖形無KDE值，則Y軸數值為不同類別分組bin間的計數值，為數據實際的離散分布狀況。
z = np.random.normal(size=100)
sns.distplot(z);
```

```python
# bin: 指的是特徵值。
#當圖形有KDE值，則Y軸數值為內核密度函數下的估計值，可視為不同類別分組bin間的相對比較值。
#當圖形無KDE值，則Y軸數值為不同類別分組bin間的計數值，為數據實際的離散分布狀況。
sns.distplot(z, bins=15, kde=False, rug=True);
```

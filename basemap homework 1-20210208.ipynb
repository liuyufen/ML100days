```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as n
```

```python
#
map = Basemap(projection='ortho', lat_0 = 0, lon_0 = 0)
#
fig = plt.figure(figsize=(8, 6),edgecolor = 'w')
#NASA                 
map.bluemarble()
#
plt.show()
```

```python
#Lambert
# width=12000000,height=9000000
#lat_1=45.,lat_2=55,lat_0=50,lon_0=-107
# projection='lcc'
#set resolution = None
#NASA  
map = Basemap(width=12000000,height=9000000,projection='lcc',resolution=None,lat_1=45.,lat_2=55,lat_0=50,lon_0=-107.)
map.bluemarble()
plt.show()
```

```python
#  "projection" Mill"
# 
#  - llcrnrlat = -90,
#  - llcrnrlon = -180,
#  - urcrnrlat = 90,
#  -urcrnrlon = 180
# resolution=c
map = Basemap(projection='mill',resolution='c',llcrnrlat = -90,llcrnrlon = -180,urcrnrlat = 90,urcrnrlon = 180)
#
map.drawcoastlines()
#2 
map.drawcountries(linewidth=2)
#
map.drawstates(color='b')
#
map.drawcounties(color='darkred')
plt.title('Basemap Tutorial')
plt.show()
```

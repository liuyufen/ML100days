```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as n
```

```python
#創建基本地圖
map = Basemap(projection='ortho', lat_0 = 0, lon_0 = 0)
#設定圖形大小
fig = plt.figure(figsize=(8, 6),edgecolor = 'w')
#地圖為NASA寶石藍的大理石版本                 
map.bluemarble()
#秀圖
plt.show()
```

```python
#設置Lambert保形底圖。
# width=12000000,height=9000000
#經緯度lat_1=45.,lat_2=55,lat_0=50,lon_0=-107
# projection='lcc'
#set resolution = None跳過邊界數據集的處理
#選用NASA 寶石藍輸出 
map = Basemap(width=12000000,height=9000000,projection='lcc',resolution=None,lat_1=45.,lat_2=55,lat_0=50,lon_0=-107.)
map.bluemarble()
plt.show()
```

```python
#  "projection" 參數為“Mill"
# 設定經緯座標
# 左下角的緯度 - llcrnrlat = -90,
# 左下角的經度 - llcrnrlon = -180,
# 右上角的緯度 - urcrnrlat = 90,
# 右上角的經度 -urcrnrlon = 180
# 成圖設定resolution=c
map = Basemap(projection='mill',resolution='c',llcrnrlat = -90,llcrnrlon = -180,urcrnrlat = 90,urcrnrlon = 180)
#創建邊線
map.drawcoastlines()
#畫出國家，並使用線寬為2 的線條生成分界線。
map.drawcountries(linewidth=2)
#畫出州界
map.drawstates(color='b')
#畫出城市
map.drawcounties(color='darkred')
plt.title('Basemap Tutorial')
plt.show()
```

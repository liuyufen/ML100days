```python
#假設丟擲的銅板是公正的，所以每次丟擲出現正面的機率為0.5，出現反面的機率為0.5。所以重複且獨立丟擲100次銅板的結果總和為二項分配。
#丟一個銅板100次，出現正面50次的機率為0.07958923738717888
p=0.5
n=100
r = 50 
probs = stats.binom.pmf(r, n, p)
print(probs)
```

```python
#假設丟擲的銅板是公正的，所以每次丟擲出現正面的機率為0.5，出現反面的機率為0.5。所以重複且獨立丟擲100次銅板的結果總和為二項分配。
#丟一個銅板100次，出現正面50次的機率為
p=0.5
n=100
r = np.arange(0,101) 
probs = stats.binom.pmf(r, n, p)
plt.bar(r, probs)
plt.ylabel('P(X=x)')
plt.xlabel('x')
plt.title('binomial(n=100,p=0.5)')
plt.show()
```
